```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
import pickle
with open('weblog.pkl', 'rb') as f:
    weblog=pickle.load(f)

len(weblog)

```

```python
import collections

site_freqs = collections.Counter()
num_rec = 0
maxlen = 0

for logs in weblog:
    if len(logs) > maxlen:
        maxlen = len(logs)
    for site in logs:
        site_freqs[site]+=1
    num_rec +=1
```

```python
site_freqs.most_common(20)
```

```python
from gensim.models import Word2Vec
EMB_SIZE = 123
```

```python
model=Word2Vec(weblog, size=EMB_SIZE, window=5, min_count=100)
```

```python
model.wv.vectors
```

```python
model.wv.most_similar('소셜커머스', topn=10)
```

```python
model.wv.most_similar(positive=['남성화전문몰','여성화전문몰'], negative=['오픈마켓'], topn=10)
```

```python
model.wv.index2word
```

```python
site_list=pd.DataFrame(model.wv.index2word, columns=['site'])
site_list[site_list['site'].str.contains('남성')]
```

```python
a= site_list[site_list['site'].str.contains('남성')].values
a = [m[0] for m in a]
a
```

```python
model.wv.most_similar(positive=a, topn=10)
```

```python
weight = model.wv.vectors
site_list = list(model.wv.vocab)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate = 100)
transformed = tsne.fit_transform(weight)
```

```python
xs = transformed[:,0]
ys = transformed[:,1]
```

```python
xs.shape
```

```python
fig = plt.figure(figsize=(30,30))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(xs, ys)

for i, word in enumerate(site_list):
    plt.annotate(word, xy=(xs[i], ys[i]))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
weight_pc = pca.fit_transform(weight)
```

```python
xs = weight_pc[:,0]
ys = weight_pc[:,1]
```

```python
fig = plt.figure(figsize=(30,30))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(xs, ys)

for i, word in enumerate(site_list):
    plt.annotate(word, xy=(xs[i], ys[i]))
```
